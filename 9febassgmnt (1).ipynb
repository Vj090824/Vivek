```python
Q.no.1: Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.



Ans: 
    

class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
        
        
In this class, the __init__ method takes three parameters: name_of_vehicle, max_speed, and average_of_vehicle. 
These parameters are used to initialize the instance variables self.name_of_vehicle, self.max_speed,
and self.average_of_vehicle.

You can create instances of the Vehicle class and access their attributes as follows:


    Create an instance of Vehicle

car = Vehicle("Car", 200, 40)

# Access the attributes of the car instance
print(car.name_of_vehicle)

Output: Car
print(car.max_speed)  

Output: 200
print(car.average_of_vehicle) 

Output: 40

In this example, we create an instance of the Vehicle class called car and pass the values "Car" for name_of_vehicle,
200 for max_speed, and 40 for average_of_vehicle. 
We can then access the attributes of the car instance using dot notation.




Q.no.2:  Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.



Ans: 
       
class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity} people."

     In  this example, the Car class is defined as a child class of the Vehicle class. 

        It inherits all the attributes and methods from the parent class.

The seating_capacity method takes a capacity parameter and returns a string that includes the name of the
vehicle and its seating capacity.

 An example of how you can create an instance of the Car class and use the seating_capacity method:


# Create an instance of Car
sedan = Car("Sedan", 180, 4)

# Use the seating_capacity method
print(sedan.seating_capacity(5))

Output:

Sedan has a seating capacity of 5 people.


In this example, we create an instance of the Car class called sedan with the name "Sedan",
a max speed of 180, and an average of 4.
We then call the seating_capacity method on the sedan instance, passing 5 as the capacity.
The method returns a string that indicates the name of the vehicle ("Sedan") and its seating capacity (5 people).




Q.no.3:  What is multiple inheritance? Write a python code to demonstrate multiple inheritance.


 
    
Ans: 
Multiple inheritance is a feature in object-oriented programming languages that allows a class to inherit attributes
and methods from more than one parent class. In multiple inheritance, a class can inherit from multiple base classes,
which means it can acquire properties and behaviors from all of its parent classes.



class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")


class Flyable:
    def fly(self):
        print(f"{self.name} is flying.")


class Bird(Animal, Flyable):
    def __init__(self, name):
        super().__init__(name)


bird = Bird("Sparrow")

bird.eat()  # Output: Sparrow is eating.
bird.sleep()  # Output: Sparrow is sleeping.
bird.fly()  # Output: Sparrow is flying.



In this example, we have three classes: Animal, Flyable, and Bird.

The Animal class represents generic attributes and behaviors of an animal.
The Flyable class represents attributes and behaviors related to flying.
The Bird class is a child class of both Animal and Flyable classes, demonstrating multiple inheritance.

The Bird class inherits from both Animal and Flyable classes using the syntax class Bird(Animal, Flyable):.
This allows the Bird class to have access to the attributes and methods defined in both parent classes.

When we create an instance of the Bird class and call its methods,
it can utilize the inherited methods from both parent classes.
In this case, the Bird instance named bird can call the eat() 
and sleep() methods inherited from the Animal class as well as
the fly() method inherited from the Flyable class.





Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.


Ans:  
     
In Python, getters and setters are methods used to access and modify the values of class attributes or properties. 
They provide a way to encapsulate the access and modification of attributes,
allowing for controlled access and data validation.

A getter method, also known as an accessor method, retrieves the value of an attribute, while a setter method, 
also known as a mutator method, sets or updates the value of an attribute. By using getter and setter methods,
we can enforce certain rules or validations when accessing or modifying the attributes of a class.



class Person: 
    def __init__(self, name, age):
        self._name = name
        self._age = age

    def get_name(self):
        return self._name

    def set_name(self, name):
        self._name = name

    def get_age(self):
        return self._age

    def set_age(self, age):
        if age < 0:
            print("Age cannot be negative.")
        else:
            self._age = age


# Create an instance of the Person class
person = Person("John", 25)

# Use getter methods
print(person.get_name())  # Output: John
print(person.get_age())  # Output: 25

# Use setter methods
person.set_name("Alice")
person.set_age(-30)  # Output: Age cannot be negative.
person.set_age(30)

# Verify the updated values using getter methods
print(person.get_name())  # Output: Alice
print(person.get_age())  # Output:30
In this example, we have a Person class with attributes _name and _age.
The class provides getter and setter methods for both attributes.

The getter methods (get_name() and get_age()) simply return the values of the attributes.

The setter methods (set_name() and set_age()) allow for modifying the attribute values.
In the case of set_age(), we have added a validation check to ensure that the age provided is not negative.

We create an instance of the Person class named person and use the getter 
methods to retrieve the initial values of the attributes.
Then, we use the setter methods to update the attribute values, including a negative value for age to demonstrate
the validation check. Finally, we use the getter methods again to verify the updated values of the attributes.






Q.no.5:   What is method overriding in python? Write a python code to demonstrate method overriding.


Ans:  

    Method overriding is a concept in object-oriented programming where a subclass provides a different implementation 
of a method that is already defined in its parent class. It allows the subclass to redefine the behavior of the method
according to its specific requirements.

When a method is overridden in the subclass, the version of the method defined in the subclass is called instead of
the version defined in the parent class when that method is invoked on an instance of the subclass.

 An example that demonstrates method overriding in Python:

class Animal:
    def sound(self):
        print("The animal makes a sound.")


class Dog(Animal):
    def sound(self):
        print("The dog barks.")


class Cat(Animal):
    def sound(self):
        print("The cat meows.")


animal = Animal()
dog = Dog()
cat = Cat()

animal.sound() 

Output: The animal makes a sound.

dog.sound() 

Output: The dog barks.
cat.sound()


Output: The cat meows.


In this example, we have a base class Animal with a method called sound(). 
The Dog and Cat classes are subclasses of Animal 
and they both override the sound() method with their own implementation.

When we create instances of Animal, Dog, and Cat and call the sound() method on each of them, the appropriate version
of the method defined in the respective class is executed.

In the example output, animal.sound() calls the method defined in the Animal class, dog.sound() calls the method defined
in the Dog class, and cat.sound() calls the method defined in the Cat class. 
This demonstrates how method overriding allows subclasses to provide their own implementation of a method defined in
the parent class.
```
