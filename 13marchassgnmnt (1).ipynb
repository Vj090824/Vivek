```python
Q1. Explain the assumptions required to use ANOVA and provide examples of violations that could impact
the validity of the results.



Ans:
    
    Analysis of Variance (ANOVA) is a statistical method used to compare means of two or more groups to determine 
    if there are significant differences between them. For ANOVA to be valid, several assumptions must be met. 
    These assumptions are essential to ensure the accuracy and reliability of the results. Let's go 
    through the key assumptions and examples of violations that can impact the validity of ANOVA results:

1. Independence of observations:
Assumption: Observations within each group should be independent of each other, meaning that the value of one
observation should not be influenced by the value of another observation in the same group.
   Violation example: In a medical study, if multiple measurements are taken from the same patient over time, 
    and these repeated measurements are used as separate data points in the ANOVA analysis,
    it violates the assumption of independence.

2. Normality:
Assumption: The data within each group should be approximately normally distributed.
   Violation example: If the data is heavily skewed or has a non-normal distribution, the ANOVA results may be unreliable. 
For instance, if a small sample size is used,the normality assumption becomes more critical, and violations can impact
    the accuracy of the analysis.

3. Homogeneity of variance (Equal variance):
Assumption: The variance of the data in each group should be approximately equal.
   Violation example: If the variance in one group is significantly larger or smaller than the variances in other groups,
    the ANOVA results can be affected. Unequal variances can lead to incorrect conclusions about the significance
    of differences between group means.

4. Random sampling:
Assumption: The data should be collected through a random sampling process to ensure 
that the findings can be generalized to the population.
   Violation example: If the data is obtained using non-random or biased sampling methods,
    the generalizability of the ANOVA results may be compromised.

5. Independence of groups:
Assumption: Different groups should be mutually exclusive and not influenced by each other.
   Violation example: If participants or subjects are part of multiple groups simultaneously,
    it can lead to dependencies and violate the independence assumption. For example,
    if the same individuals are included in multiple treatment groups in a study.

6. Interval or ratio-level data:
Assumption: The dependent variable should be measured on an interval or ratio scale.
   Violation example: If the dependent variable is measured on a nominal or ordinal scale, ANOVA may not be appropriate,
    and other non-parametric tests should be considered.

If any of these assumptions are violated, the results of the ANOVA may not be valid, and the conclusions drawn from 
the analysis may be unreliable. It's crucial to check for these assumptions before conducting 
an ANOVA and consider alternative statistical tests if the assumptions are not met. Additionally,
data transformation techniques or robust ANOVA methods can sometimes help address violations, 
but they should be used with caution and based on the specific nature of the data and the research question.










Q2. What are the three types of ANOVA, and in what situations would each be used?


Ans:
      
        
Apologies for the confusion. It seems my previous response had the same question and answer as the previous one. 
Let me provide you with the correct response:

The three types of ANOVA are:

1. One-Way ANOVA:
   One-Way ANOVA is used when you have one categorical independent variable (also known as a factor) 
with three or more levels or groups, and you want to compare the means of the dependent variable among these groups. 
It is used to determine if there are significant differences between the means of multiple groups.

   Example: A researcher wants to compare the test scores (dependent variable) of students from three different schools
(Group A, Group B, and Group C). One-Way ANOVA would be used to determine if there are significant differences in test 
scores between the three schools.

2. Two-Way ANOVA:
   Two-Way ANOVA is used when you have two categorical independent variables (factors), and you want to examine their
main effects and interaction effect on the dependent variable. Each independent variable has two or more levels, 
and the dependent variable is continuous.

   Example: A scientist wants to investigate the effects of two factors, temperature (high and low)
    and fertilizer type (A and B), on plant growth (dependent variable).
    Two-Way ANOVA would be used to determine if there are significant main effects of temperature 
    and fertilizer type on plant growth and if there is any interaction between these two factors.

3. Repeated Measures ANOVA:
   Repeated Measures ANOVA is used when you have one group of participants or subjects, 
and you measure the same dependent variable multiple times under different conditions or at different time points.
This type of ANOVA allows you to test for within-subject differences across the various time points or conditions.

   Example: A psychologist wants to investigate the effectiveness of a new therapy for anxiety. 
    They measure anxiety levels (dependent variable) in the same group of participants before the therapy,
    after one week of therapy, and after two weeks of therapy. Repeated Measures ANOVA would be used 
    to determine if there are significant differences in anxiety levels across the three time points.

Each type of ANOVA is appropriate for different study designs and research questions. It is essential
to choose the correct type of ANOVA based on the number of independent variables, the design of the study,
and the nature of the data collected.









Q3. What is the partitioning of variance in ANOVA, and why is it important to understand this concept?



Ans:
    
    In Analysis of Variance (ANOVA), the partitioning of variance refers to the process of decomposing the total
        variation observed in a dataset into different sources or components of variation.
        ANOVA is a statistical technique used to compare the means of three or more groups to determine
        if there are any statistically significant differences among them.

The variance is a measure of the dispersion or spread of data points in a dataset.
In the context of ANOVA, the total variance is the overall variability observed in the dependent variable 
(the one being measured) across all groups. By partitioning this total variance into different components,
ANOVA allows us to understand how much of the variability in the 
dependent variable can be attributed to different factors.

There are three main components in the partitioning of variance in a one-way ANOVA:

1. Between-group variance (explained variance): This component represents the variability in 
the dependent variable that can be attributed to the differences between the group means. 
It indicates how much of the total variance is explained by the grouping variable 
(the factor or independent variable under study).

2. Within-group variance (unexplained variance or error variance):
    This component represents the variability in the dependent variable that cannot be accounted
    for by the group means. It reflects the random variability or error within each group,
    which is not explained by the grouping variable.

3. Total variance: This is the sum of the between-group variance and the within-group variance, 
representing the overall variability in the dependent variable.

The importance of understanding the partitioning of variance in ANOVA lies in the insights it provides:

1. Identifying significant differences: By comparing the between-group variance to the within-group variance,
ANOVA helps determine if there are significant differences between the group means. 
If the between-group variance is significantly larger than the within-group variance,
it suggests that the groups are not equal, and there may be a significant effect of the independent variable.

2. Assessing the importance of factors: ANOVA allows researchers to assess the relative contribution
of different factors to the overall variation in the dependent variable. 
This helps to identify which factors have the most substantial influence and are worth further investigation.

3. Drawing conclusions: Understanding the partitioning of variance helps researchers draw appropriate conclusions
about the relationships between variables. It provides a more nuanced understanding of the data than
simply comparing group means, as it takes into account the variability within the groups.

4. Designing effective experiments: For experimental design, knowing how to partition variance can guide researchers 
in determining the appropriate sample sizes and controlling for potential confounding variables, 
thus increasing the validity and reliability of their findings.

Overall, the partitioning of variance in ANOVA is a fundamental concept that helps researchers analyze 
and interpret the results of their experiments accurately, 
leading to more informed conclusions and better scientific understanding.









Q4. How would you calculate the total sum of squares (SST), explained sum of squares (SSE), and residual
sum of squares (SSR) in a one-way ANOVA using Python?
          
    
Ans:
    
    
    In a one-way ANOVA (Analysis of Variance), we aim to partition the total variance of the data into two components:
        the variance explained by the model and the variance left unexplained (residual variance). 
We can calculate the Total Sum of Squares (SST), Explained Sum of Squares (SSE),
and Residual Sum of Squares (SSR) using Python.

Let's assume we have a dataset with a dependent variable and a categorical independent 
variable (the factor or group). We want to compare the means of the dependent variable across different groups.

To calculate SST, SSE, and SSR, we can use the following formulas:

Total Sum of Squares (SST):
SST represents the total variance in the dependent variable. It is calculated as the sum of the
squared differences between each data point and the overall mean of the dependent variable.

Explained Sum of Squares (SSE):
SSE represents the variance in the dependent variable that can be explained by the model (group means).
It is calculated as the sum of the squared differences between each group mean and the overall mean of
the dependent variable, weighted by the number of observations in each group.

Residual Sum of Squares (SSR):
SSR represents the unexplained variance in the dependent variable. It is calculated as the sum of the 
squared differences between each data point and its corresponding group mean.





First, we'll need to have your data organized in different groups,
and then you can follow these steps:

Import the required libraries:

import numpy as np
from scipy.stats import f_oneway
Prepare your data:
    
    
Let's assume you have multiple groups with their corresponding data stored in separate arrays or lists.


group1_data = [10, 15, 20, 18, 22]
group2_data = [25, 30, 28, 32, 35]
group3_data = [40, 45, 38, 42, 50]
# Add more groups if you have them
Calculate the means and overall mean:

# Calculate the means of each group
mean_group1 = np.mean(group1_data)
mean_group2 = np.mean(group2_data)
mean_group3 = np.mean(group3_data)
# Add more means for additional groups

# Calculate the overall mean
overall_mean = np.mean([mean_group1, mean_group2, mean_group3])
Calculate the sum of squares:

    
# Calculate the Total Sum of Squares (SST)
sst = sum((np.array(group1_data) - overall_mean) ** 2) + \
      sum((np.array(group2_data) - overall_mean) ** 2) + \
      sum((np.array(group3_data) - overall_mean) ** 2)
# Add more sums for additional groups

# Calculate the Explained Sum of Squares (SSE)
sse = len(group1_data) * (mean_group1 - overall_mean) ** 2 + \
      len(group2_data) * (mean_group2 - overall_mean) ** 2 + \
      len(group3_data) * (mean_group3 - overall_mean) ** 2
# Add more sums for additional groups

# Calculate the Residual Sum of Squares (SSR)
ssr = sst - sse
Perform the one-way ANOVA and verify the results:

    

# Perform the one-way ANOVA test
f_statistic, p_value = f_oneway(group1_data, group2_data, group3_data)
# Add more groups if you have them

# Print the results
print("Total Sum of Squares (SST):", sst)
print("Explained Sum of Squares (SSE):", sse)
print("Residual Sum of Squares (SSR):", ssr)
print("F-statistic:", f_statistic)
print("p-value:", p_value)
The F-statistic and p-value obtained from the one-way ANOVA test can help you determine whether there are
significant differences between the group means. A small p-value (typically less than 0.05)
would indicate that at least one group mean is significantly different from the others.









Q5. In a two-way ANOVA, how would you calculate the main effects and interaction effects using Python?


Ans: 
        
        
In a two-way ANOVA (Analysis of Variance), we have two categorical independent variables (factors)
and one continuous dependent variable. The main effects represent the individual effects of each factor,
and the interaction effect represents how the combination of the two factors influences the dependent variable.

To calculate the main effects and interaction effects using Python, we can use the `statsmodels` library,
which provides a comprehensive set of tools for statistical analysis. 
First, make sure you have `statsmodels` installed.
You can install it using pip if you haven't already:


pip install statsmodels


Next, you'll need to have your data ready in a pandas DataFrame, with the dependent variable in one column 
and the two categorical factors in separate columns. Let's assume you have the following DataFrame named `data`:


import pandas as pd

# Sample data with two factors (A and B) and a dependent variable (DV)
data = pd.DataFrame({
    'A': [1, 2, 1, 2, 1, 2, 1, 2],
    'B': [1, 1, 2, 2, 1, 1, 2, 2],
    'DV': [10, 15, 12, 18, 8, 11, 9, 14]
})


Now, you can perform the two-way ANOVA and calculate the main effects and interaction effects as follows:


import statsmodels.api as sm
from statsmodels.formula.api import ols

# Perform two-way ANOVA
model = ols('DV ~ A + B + A:B', data=data).fit()
anova_table = sm.stats.anova_lm(model)

# Extract main effects and interaction effect
main_effect_A = anova_table.loc['A', 'sum_sq'] / anova_table.loc['A', 'df']
main_effect_B = anova_table.loc['B', 'sum_sq'] / anova_table.loc['B', 'df']
interaction_effect = anova_table.loc['A:B', 'sum_sq'] / anova_table.loc['A:B', 'df']

# Display the results
print("Main effect of A:", main_effect_A)
print("Main effect of B:", main_effect_B)
print("Interaction effect:", interaction_effect)


The `anova_lm()` function from `statsmodels` performs the ANOVA, and the results are stored in `anova_table`. 
We then extract the sum of squares for each factor and divide it by the respective degrees of freedom to obtain
the main effects and interaction effect.











Q6. Suppose you conducted a one-way ANOVA and obtained an F-statistic of 5.23 and a p-value of 0.02.
What can you conclude about the differences between the groups, and how would you interpret these
results?




Ans:
      
In a one-way ANOVA, we test whether there are significant differences in
the means of three or more independent groups. 

The F-statistic and the associated p-value are used to make conclusions
about the differences between the groups. In your case:

- F-statistic: 5.23
- p-value: 0.02

To interpret these results:

1. F-statistic: The F-statistic is a measure of the variability between the group means compared 
to the variability within the groups. A higher F-statistic suggests that there are greater differences between
the group means relative to the variability within each group.

2. p-value: The p-value is the probability of obtaining results as extreme as the observed data,
assuming that the null hypothesis is true. In the context of a one-way ANOVA, the null hypothesis states
that there are no significant differences between the group means.

Given the p-value of 0.02, which is less than the commonly chosen significance level of 0.05, 
we can conclude that there are statistically significant differences between the groups. 
In other words, the data provides strong evidence to reject the null hypothesis,
suggesting that at least one group mean is significantly different from the others.

However, the p-value alone does not indicate which specific group means are different from each other.
To determine that, you might conduct post-hoc tests (e.g., Tukey's HSD or Bonferroni correction)
to make pairwise comparisons between the groups and identify which pairs have significant differences in means.

In summary, the F-statistic indicates that there are differences in means among the groups,
    and the low p-value indicates that these differences are statistically significant. 
 Further analysis with appropriate post-hoc tests would be necessary to pinpoint the specific group differences.
                                                     
                                                     
                                                     
                                                     
                                                     
                                                     
                                                     
                                                     
 
                                                     
                                                     
                                                     
  Q7. In a repeated measures ANOVA, how would you handle missing data, and what are the potential
consequences of using different methods to handle missing data?
                                                     
                                                     
                                                     
                                                     
 Ans:
         Handling missing data in a repeated measures ANOVA is crucial to obtain valid and reliable results.
There are several methods to deal with missing data, and the choice of method can influence the outcomes 
and interpretations of the analysis. Here are some common approaches to handle missing 
        data and their potential consequences:

1. Complete Case Analysis (Listwise Deletion):
   - This method involves excluding all cases with missing data for any variable involved in the analysis.
   - Consequence: This approach can lead to biased estimates and loss of power if the missing data are 
not missing completely at random (MCAR). Additionally, it reduces the sample size and may not 
be efficient if the amount of missing data is substantial.

2. Mean/Mode Imputation:
   - In this method, missing values for a variable are replaced by the mean (for continuous data)
    or mode (for categorical data) of that variable.
   - Consequence: Mean or mode imputation may underestimate the standard errors and lead to inflated 
Type I error rates. It also reduces the variability in the data and does not capture the 
uncertainty introduced by imputing the missing values.

3. Last Observation Carried Forward (LOCF):
   - This approach involves using the last observed value for a participant for any
    subsequent missing data points.
   - Consequence: LOCF can introduce bias if the reason for missingness is related to the individual's trajectory.
It may also not accurately represent the true values for the missing time points.

4. Multiple Imputation:
   - Multiple imputation creates several plausible imputed datasets, each with different imputed values 
    for the missing data, based on the observed data and certain assumptions.
   - Consequence: Multiple imputation accounts for the uncertainty associated with missing data
and provides more accurate estimates compared to single imputation methods.
                                                     
However, it requires appropriate handling of the imputation model and could be computationally intensive.

5. Maximum Likelihood Estimation (MLE):
                                                     
   - MLE estimates model parameters by maximizing the likelihood function that incorporates information 
        from all available data, including the cases with missing data.
   - Consequence: MLE is a powerful approach when data are missing at random (MAR) or MCAR. 
However, it may yield biased results if the data are not MAR or if the assumptions underlying the model are violated.

It's essential to carefully consider the nature of the missing data and select an appropriate 
method that aligns with the underlying assumptions. 
Multiple imputation and maximum likelihood estimation are generally considered more robust and preferable
when dealing with missing data. Additionally, sensitivity analyses can be conducted to assess 
the impact of different missing data methods on the results and conclusions of the repeated measures ANOVA.     
                                                     
                                                     
                                                     
                                                     
                                                     
                                                     
                                                     
  
     
                                                     
Q8. What are some common post-hoc tests used after ANOVA, and when would you use each one? Provide
an example of a situation where a post-hoc test might be necessary.   
                                                     
                                                     
                                                     
                                                     
 Ans:
                                                     
       After conducting an ANOVA (Analysis of Variance) and finding a significant difference among means,
    post-hoc tests are used to determine which specific groups differ from each other.
Post-hoc tests are essential because ANOVA only tells us that there is a significant difference
    somewhere, but it doesn't specify which groups are significantly different.
            Some common post-hoc tests include:

1. Tukey's Honestly Significant Difference (Tukey's HSD): 
Tukey's HSD test is widely used and compares all possible pairs of group means 
to determine significant differences. It controls the familywise error rate, making it suitable for 
situations where you have several groups to compare.

2. Bonferroni correction: Bonferroni is a conservative post-hoc test that adjusts the significance level to maintain
a desired familywise error rate. It divides the desired significance level (alpha) by the number of comparisons being made.

3. Scheffe's method: Scheffe's method is less restrictive than Tukey's HSD and Bonferroni, 
making it useful when dealing with unequal sample sizes or complex comparisons.

4. Dunn's test: Dunn's test is a non-parametric post-hoc test used when ANOVA assumptions are violated,
or data is not normally distributed.

Example of a situation where a post-hoc test might be necessary:

Let's say a researcher conducts an experiment to compare the effectiveness of three different training methods
(A, B, and C) on improving employees' performance. The dependent variable is the score achieved on
a performance assessment test. After conducting a one-way ANOVA, 
the researcher finds a significant difference in mean scores among the training groups (p < 0.05).

Now, the researcher needs to use a post-hoc test to determine which specific training methods are
significantly different from each other. For this example, they decide to use Tukey's HSD. 
After running Tukey's test, the researcher finds that Group A and Group B have a significantly
higher mean score compared to Group C. However, there is no significant difference between Group A and Group B.

In this situation, the post-hoc test (Tukey's HSD) was necessary to identify the specific groups 
that differ significantly in performance. Without the post-hoc test, the researcher would have only
known that there is a significant difference among the training methods but would not have been able to 
pinpoint which groups were responsible for the difference.                           
                                      
                                      
                                      

                                      
                                      

                                      
                                      
                                      
 Q9. A researcher wants to compare the mean weight loss of three diets: A, B, and C. They collect data from
50 participants who were randomly assigned to one of the diets. Conduct a one-way ANOVA using Python
to determine if there are any significant differences between the mean weight loss of the three diets.
Report the F-statistic and p-value, and interpret the results.


                                      
                                      
                                      
 
 Ans:
                                      
       To conduct a one-way ANOVA in Python, we can use the `scipy.stats` module. 
    First, make sure you have the necessary libraries installed:


pip install scipy numpy

Next, let's assume you have the weight loss data for each diet (A, B, and C) in three separate lists:
    `weight_loss_A`, `weight_loss_B`, and `weight_loss_C`. Each list should contain the weight
    loss values for the 50 participants assigned to the respective diets.

Here's how you can perform the one-way ANOVA and interpret the results:


import numpy as np
from scipy.stats import f_oneway

# Weight loss data for each diet
weight_loss_A = [3.2, 2.9, 4.1, 3.8, 3.5, ...]  # Replace the ellipsis with the actual data

weight_loss_B = [2.1, 2.5, 1.9, 2.8, 2.3, ...]  # Replace the ellipsis with the actual data
                             
weight_loss_C = [1.5, 1.7, 1.9, 2.0, 1.4, ...]  # Replace the ellipsis with the actual data

# Perform one-way ANOVA
F_statistic, p_value = f_oneway(weight_loss_A, weight_loss_B, weight_loss_C)

# Print the results
print("F-statistic:", F_statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05  # Set the significance level

if p_value < alpha:
    print("There is a significant difference between the mean weight loss of the three diets.")
else:
    print("There is no significant difference between the mean weight loss of the three diets.")


In this code, we first import the necessary libraries (`numpy` and `scipy.stats`).
    Then, we define the weight loss data for each diet as separate lists. 
We use the `f_oneway` function from `scipy.stats` to perform the one-way ANOVA, 
    which calculates the F-statistic and p-value.

Finally, we interpret the results by comparing the p-value with the chosen significance level (alpha). 
If the p-value is less than alpha (0.05 in this case), we reject the null hypothesis, 
indicating that there is a significant difference between the mean weight loss of the three diets.
If the p-value is greater than or equal to alpha, we fail to reject the null hypothesis,
suggesting that there is no significant difference between the mean weight loss of the three diets.          
                                      
                                      
                                      
  
                                      
                                      
                                      
                                      
                                      
                                      
   
  Q10. A company wants to know if there are any significant differences in the average time it takes to
complete a task using three different software programs: Program A, Program B, and Program C. They
randomly assign 30 employees to one of the programs and record the time it takes each employee to
complete the task. Conduct a two-way ANOVA using Python to determine if there are any main effects or
interaction effects between the software programs and employee experience level (novice vs.
experienced). Report the F-statistics and p-values, and interpret the results.
                                      
                                      
                              
                                      
Ans:
                                      
  To conduct a two-way ANOVA using Python, we can use the `statsmodels` library,
which provides a wide range of statistical functions.
    First, make sure you have `statsmodels` and other required libraries installed:


pip install numpy pandas statsmodels


Now, let's assume you have the data in a DataFrame called `data` with three columns:
"Software_Program" (Program A, Program B, or Program C), "Experience_Level" (novice or experienced),
and "Completion_Time" (time taken to complete the task).
You'll need to import the necessary libraries and run the ANOVA:


import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Sample data in the 'data' DataFrame

# Perform the two-way ANOVA
model = ols('Completion_Time ~ Software_Program + Experience_Level +
Software_Program:Experience_Level', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Extract F-statistics and p-values
F_software = anova_table.loc['Software_Program', 'F']
p_software = anova_table.loc['Software_Program', 'PR(>F)']

F_experience = anova_table.loc['Experience_Level', 'F']
p_experience = anova_table.loc['Experience_Level', 'PR(>F)']

F_interaction = anova_table.loc['Software_Program:Experience_Level', 'F']
p_interaction = anova_table.loc['Software_Program:Experience_Level', 'PR(>F)']

# Print the results
print("Software Program F-statistic:", F_software)
print("Software Program p-value:", p_software)

print("Experience Level F-statistic:", F_experience)
print("Experience Level p-value:", p_experience)

print("Interaction F-statistic:", F_interaction)
print("Interaction p-value:", p_interaction)


Interpretation of results:
- F-statistic and p-value for Software Program:
The F-statistic tests the main effect of the software program on task completion time. 
The p-value determines the statistical significance of this effect. 
A small p-value (typically less than 0.05) suggests that there is a significant difference
in the average time it takes to complete the task across the three software programs.

- F-statistic and p-value for Experience Level:
The F-statistic tests the main effect of employee experience level on task completion time. 
The p-value indicates the significance of this effect. 
A low p-value suggests that there is a significant difference in the average time taken to
complete the task between novice and experienced employees.

- F-statistic and p-value for Interaction:
The F-statistic tests the interaction effect between software programs and employee experience levels. 
The interaction effect assesses whether the effect of one factor 
(e.g., software program) on the response variable (completion time) depends on the levels of another factor
(e.g., experience level). A small p-value indicates that there is a significant interaction effect, 
meaning the impact of software programs on completion time differs based on employee experience level.

Remember that p-values below the significance level (often 0.05) suggest evidence for significant effects. 

            
            
            
 
            
            

            
 Q11. An educational researcher is interested in whether a new teaching method improves student test
scores. They randomly assign 100 students to either the control group (traditional teaching method) or the
experimental group (new teaching method) and administer a test at the end of the semester. Conduct a
two-sample t-test using Python to determine if there are any significant differences in test scores
between the two groups. If the results are significant, follow up with a post-hoc test to determine which
group(s) differ significantly from each other.
            
            
Ans:           
    


To conduct a two-sample t-test and post-hoc test in Python,
    we'll use the `scipy` library for the t-test and `statsmodels` 
for the post-hoc test. First, make sure you have these libraries installed. 
If you don't, you can install them using `pip`:


pip install scipy
pip install statsmodels


Next, let's assume you have the test scores for the control and experimental
groups stored in two separate lists or arrays.
For this example, I'll create random data for demonstration purposes.
You can replace these with your actual data:


import numpy as np
from scipy import stats
import statsmodels.stats.multicomp as mc

# Random test scores for control and experimental groups (replace with actual data)
np.random.seed(42)  # For reproducibility
control_group_scores = np.random.normal(loc=70, scale=10, size=100)
experimental_group_scores = np.random.normal(loc=75, scale=12, size=100)

# Two-sample t-test
t_stat, p_value = stats.ttest_ind(control_group_scores, experimental_group_scores)

print("Two-sample t-test results:")
print("t-statistic:", t_stat)
print("p-value:", p_value)

# Post-hoc test (Tukey's HSD) if the results are significant
if p_value < 0.05:
    # Combine the scores and group labels
all_scores = np.concatenate([control_group_scores, experimental_group_scores])
group_labels = ['Control'] * len(control_group_scores) + ['Experimental'] * len(experimental_group_scores)
    
    # Perform Tukey's HSD post-hoc test
    posthoc_results = mc.pairwise_tukeyhsd(all_scores, group_labels)
    print("\nPost-hoc test results:")
    print(posthoc_results)



In this example, I used the `numpy` library to generate random test scores for demonstration purposes.
Replace `control_group_scores` and `experimental_group_scores` with your actual data.

The two-sample t-test will give you the t-statistic and the p-value. If the p-value is less
than your chosen significance level (e.g., 0.05), then you have evidence to suggest that there
is a significant difference in test scores between the two groups.

If the results are significant, we can follow up with a post-hoc test to determine 
which group(s) differ significantly from each other. For this example, 
I used Tukey's Honestly Significant Difference (HSD) test as a post-hoc test. 
The post-hoc test helps identify specific group pairs with significant differences in test scores.


            
            
            
            
            
            
            
            
            
Q12. A researcher wants to know if there are any significant differences in the average daily sales of three
retail stores: Store A, Store B, and Store C. They randomly select 30 days and record the sales for each store
on those days. Conduct a repeated measures ANOVA using Python to determine if there are any
significant differences in sales between the three stores.
If the results are significant, follow up with a posthoc test to determine which store(s)
differ significantly from each other.            
            
            
            
 Ans:
            
  To conduct a repeated measures ANOVA in Python and follow up with a posthoc test, we can use the `statsmodels`
library for the ANOVA and the `pingouin` library for the posthoc test. 
First, make sure you have these libraries installed in your Python environment.
You can install them using `pip` if you haven't already:


pip install statsmodels pingouin


Next, let's assume you have the data for Store A, Store B, and Store C sales for the 30 selected days
in three separate lists: `store_a_sales`, `store_b_sales`, and `store_c_sales`.

Here's a step-by-step guide on how to perform the repeated measures ANOVA and posthoc test:


import numpy as np
import pandas as pd
import pingouin as pg
from statsmodels.stats.anova import AnovaRM

# Create a DataFrame to hold the data
data = pd.DataFrame({
    'Store': ['A'] * 30 + ['B'] * 30 + ['C'] * 30,
    'Sales': store_a_sales + store_b_sales + store_c_sales
})

# Perform the repeated measures ANOVA
# "Subject" will be our grouping variable (Store), and "Sales" is the dependent variable
aov = AnovaRM(data, 'Sales', 'Store', within=['Subject'])
result_anova = aov.fit()

# Print the ANOVA results
print(result_anova)

# Follow up with a posthoc test (Tukey's HSD)
posthoc = pg.pairwise_ttests(dv='Sales', within='Store', subject='Subject',
data=data, parametric=True, padjust='bonf')
print(posthoc)


The `AnovaRM` function from `statsmodels` performs the repeated measures ANOVA, taking the data,
the dependent variable ('Sales'), the grouping variable ('Store'), and the within-subject variable
('Subject') as inputs. The `pingouin` library provides the `pairwise_ttests` function to perform the posthoc test,
in this case, Tukey's HSD (honestly significant difference) test.

The output will provide the results of the repeated measures ANOVA, including F-value,
p-value, and degrees of freedom, as well as the posthoc test results with p-values 
and significant differences between the stores.

Remember, the ANOVA assumes that the data meet the assumptions of normality, homogeneity of variance,
and independence of observations.
If these assumptions are not met, alternative tests or transformations might be necessary.
Additionally, the sample size of 30 days for each store might be relatively small.

            
```
