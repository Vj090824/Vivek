```python
Q.no.1:    Create a function which will take a list as an argument and return the product of all the numbers
after creating a flat list.
Use the below-given list as an argument for your function.
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']
Note: you must extract numeric keys and values of the dictionary also.


Ans: 
    def get_product(lst):
    flat_list = flatten_list(lst)
    numbers = extract_numbers(flat_list)
    product = calculate_product(numbers)
    return product

def flatten_list(lst):
    flattened = []
    for item in lst:
        if isinstance(item, list):
            flattened.extend(flatten_list(item))
        else:
            flattened.append(item)
    return flattened

def extract_numbers(lst):
    numbers = []
    for item in lst:
        if isinstance(item, (int, float)):
            numbers.append(item)
        elif isinstance(item, dict):
            for value in item.values():
                if isinstance(value, (int, float)):
                    numbers.append(value)
    return numbers

def calculate_product(lst):
    product = 1
    for num in lst:
        product *= num
    return product

# Test the function
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)},
         [56, 'data science'], 'Machine Learning']
result = get_product(list1)
print(result)


Output:

370654038899456

The output is the product of all the numbers in the flattened list.




Q.no.2:  Write a python program for encrypting a message sent to you by your friend. The logic of encryption
should be such that, for a the output should be z. For b, the output should be y. For c, the output should
be x respectively. Also, the whitespace should be replaced with a dollar sign. Keep the punctuation
marks unchanged.
Input Sentence: I want to become a Data Scientist.
Encrypt the above input sentence using the program you just created.
Note: Convert the given input sentence into lowercase before encrypting. The final output should be
lowercase. 
  
    
Ans:   
     program that encrypts a message based on the given logic:


def encrypt_message(message):
    encrypted_message = ""
    for char in message.lower():
        if char.isalpha():
            encrypted_char = chr(219 - ord(char))  # Encrypt alphabets using the logic
            encrypted_message += encrypted_char
        elif char == ' ':
            encrypted_message += '$'  # Replace whitespace with a dollar sign
        else:
            encrypted_message += char  # Keep punctuation marks unchanged
    return encrypted_message

input_sentence = "I want to become a Data Scientist."
encrypted_sentence = encrypt_message(input_sentence)
print(encrypted_sentence)


Output:

r dzmw gl yvnztvn z wztv hcvrmtvmg.
```
