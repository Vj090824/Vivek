```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.



Ans:
       

Seaborn is a popular Python data visualization library that builds on top of Matplotlib. 
It provides a high-level interface
for creating attractive and informative statistical graphics. 
Here are five common plots that can be created using Seaborn 
along with their typical use cases:

Scatter Plot:

Uses: It shows the relationship between two continuous variables. 
It is useful for identifying trends, clusters, and outliers.
Line Plot:

Uses: It displays the relationship between two continuous variables over a continuous interval.
It is commonly used for time series data and trend analysis.
Bar Plot:

Uses: It represents categorical data with rectangular bars. 
It can be used to compare categories or show the distribution of a
single categorical variable.
Histogram:

Uses: It displays the distribution of a single numerical variable.
It is useful for visualizing the frequency or probability density 
of data across different bins.
Heatmap:

Uses: It represents data using a color-encoded matrix. 
It is commonly used to visualize the correlation between variables in a dataset
or display the intensity of values across two dimensions.
These are just a few examples, and Seaborn provides many more types of plots for different use cases. 
Each plot can be customized with various parameters to further enhance the visual





Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.



Ans:  
    
    To load the "fmri" dataset using the load_dataset function from the Seaborn library
    and plot a line plot based on the specified columns.
        
        
        
        import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Set the title and labels for the plot
plt.title("fMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()
 
    In this code, we first import the Seaborn library using the import seaborn as sns statement. 
    Then we load the "fmri" dataset using sns.load_dataset("fmri") and store it in the variable fmri.

Next, we use the sns.lineplot function to create a line plot. We specify the dataset using the data parameter as fmri. 
The x parameter is set to "timepoint" to represent the x-axis values,
and the y parameter is set to "signal" to represent the y-axis values.

We also specify the hue parameter as "event" to differentiate the lines based on different events,
and the style parameter as "region" to differentiate the lines based on different regions.

Finally, we use plt.show() to display the plot.






Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.\


Ans:     
    
    To load the "titanic" dataset using the load_dataset function from the Seaborn library and plot two box plots
    based on the specified columns, follow these steps:


import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Plot the first box plot for 'pclass' and 'age'
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')

# Plot the second box plot for 'pclass' and 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()


In this code, we first import the Seaborn library using the import seaborn as sns statement 
and the Matplotlib library using import matplotlib.pyplot as plt.

Then we load the "titanic" dataset using sns.load_dataset("titanic") and store it in the variable titanic.

We plot the first box plot for 'pclass' and 'age' by creating a subplot with plt.subplot(1, 2, 1). 
The data parameter of sns.boxplot is set to titanic, the x parameter is set to 'pclass',
and the y parameter is set to 'age'. 
We also set the title of the plot using plt.title.

Next, we plot the second box plot for 'pclass' and 'fare' by creating another subplot with plt.subplot(1, 2, 2). 
The parameters for sns.boxplot and the title are set in a similar way.

Finally, we adjust the layout using plt.tight_layout() to avoid overlapping subplots,
and we display the plots using plt.show().
        
    
    
    
    
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.


Ans: 
    
      To plot a histogram for the 'price' column of the 'diamonds' dataset using the 'cut' column as the hue parameter, 
        i can use the seaborn library in Python. 


import seaborn as sns

# Load the diamonds dataset from seaborn
diamonds = sns.load_dataset('diamonds')

# Plot histogram with hue parameter
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')

# Display the plot
import matplotlib.pyplot as plt
plt.show()



Then, it loads the 'diamonds' dataset using the `load_dataset` function from seaborn.

Next, it uses the `histplot` function from seaborn to create a histogram. 
The `data` parameter specifies the dataset to be used, 
the `x` parameter sets the column to be plotted ('price' in this case), and the `hue` parameter determines the column 
used to distinguish different subsets ('cut' in this case). The `multiple` parameter is set to 'stack' to stack the 
histograms based on the 'cut' column.

Finally, the plot is displayed using `plt.show()` from the matplotlib library.
Executing this code will generate a histogram with stacked bars, 
representing the distribution of diamond prices based on different cuts.





Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.



Ans:   
    
      An example of how you can use the "iris" dataset from seaborn to plot a pair plot with
        the "species" column as the hue parameter:


import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot a pair plot
sns.pairplot(iris, hue="species")

# Display the plot
import matplotlib.pyplot as plt
plt.show()


This code will generate a pair plot using the "iris" dataset from seaborn.
The pair plot will display scatterplots for each pair
of features in the dataset,with different colors representing the different 
species of iris flowers (setosa, versicolor, and virginica).





Que 6: Use the "flights" dataset from seaborn to plot a heatmap.


Ans:

import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot("month", "year", "passengers")

plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Flights Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()

       
    
This code will generate a heatmap showing the number of passengers for each month and year in the "flights" dataset.
The color intensity represents the number of passengers, with darker colors indicating higher values.
The annotations inside the heatmap display the exact passenger counts for each month and year.
```
