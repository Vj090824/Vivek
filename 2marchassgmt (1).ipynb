```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.


Ans:  
      Matplotlib is a popular data visualization library for Python.
        It provides a wide range of tools and functions to create static,
        animated, and interactive plots, graphs, and charts. Matplotlib is widely used in various
        fields such as data analysis, scientific 
        research, engineering, finance, and more.

Here are five plots that can be created using the Pyplot module of Matplotlib:

Line Plot: This is a basic plot that displays data points connected by straight line segments. 
It is commonly used to show the trend or 
relationship between two variables over a continuous range.

Scatter Plot: Scatter plots are used to visualize the relationship between two variables by 
representing data points as individual dots.
Each dot represents the values of the two variables on the x and y-axis, making it useful for
identifying patterns or correlations.

Bar Plot: A bar plot represents data using rectangular bars with heights or lengths proportional
to the values they represent. 
It is often used to compare and visualize categorical or discrete data.

Histogram: Histograms provide a visual representation of the distribution of a dataset. 
It divides the data into bins and displays 
the frequency or count of values falling within each bin.
Histograms are useful for understanding the shape, central tendency,
and spread of numerical data.

Pie Chart: Pie charts are circular plots divided into slices, 
where each slice represents a proportion or percentage of a whole. 
They are commonly used to show the composition or distribution of categorical data,
such as market share, population breakdowns, etc.

These are just a few examples of the types of plots that can be created using the Pyplot module of Matplotlib. 
It offers many more plot types and customization options to cater to various visualization needs.





Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.



Ans: 
    
    A scatter plot is a type of plot used to visualize the relationship between two variables.
        It represents data points as individual dots on a graph, where each dot corresponds to a pair of values 
        from the two variables being compared.

To generate the scatter plot using the given code, you can use the matplotlib.pyplot module.
An example of how you can create the scatter plot with the provided data and add a title, xlabel, and ylabel:


import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create the scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()

This code will generate a scatter plot with the generated x and y data.
The scatter() function is used to create the scatter plot,
and the title(), xlabel(), and ylabel() functions are used to add the title, x-axis label,
and y-axis label to the plot, respectively. Finally, plt.show() is called to display the plot.




Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])



Ans: 
       The subplot() function in Matplotlib is used to create multiple subplots within a single figure. 
        It allows you to divide the figure into a grid of rows and columns and specify the position of each subplot.
    This is useful when you want to display multiple plots or visualizations side by side for comparison or
    to present different aspects of the data.

four line plots using the subplot() function with the provided data, you can use the following code:

import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()

In this code, we first define the x and y data for each line plot. 
Then, using subplots(2, 2), we create a figure with a grid of 2 rows and 2 columns to accommodate four subplots.
Each subplot is accessed using indexing with axs[row_index, column_index].

We plot each line using plot() on the respective subplot and set the title for each subplot using set_title(). 
Finally, tight_layout() is used to adjust the spacing between subplots, and show() displays the plot.





Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])




Ans:
       
    A bar plot is a type of data visualization that uses rectangular bars to represent data. 
    It is commonly used to display and compare categorical data or to show the distribution
    of a numerical variable across different categories. 
    The length of each bar corresponds to the value of the variable being plotted.

Bar plots are used for several purposes:

Comparing values: Bar plots allow for easy visual comparison between different categories or groups.
By comparing the lengths of the bars, 
you can quickly identify which categories have higher or lower values.

Showing trends: Bar plots can be used to visualize trends over time or across different conditions.
By plotting the values on the y-axis
and categories on the x-axis, you can observe how the values change or differ for each category.

Presenting frequencies or counts: Bar plots are useful for displaying the frequency or 
count of occurrences of different categories. 
The height of each bar represents the frequency or count of the corresponding category.

Now, create a bar plot and a horizontal bar plot using the given data:

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Profit Comparison")
plt.show()

# Create a horizontal bar plot

plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Profit Comparison")
plt.show()

The first plot will be a bar plot with the company names on the x-axis and the corresponding profits on the y-axis. 
 The second plot will be a horizontal bar plot with the company names on the y-axis and the profits on the x-axis.




Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)




Ans: 
    
      A box plot, also known as a box-and-whisker plot, is a graphical representation 
        of a dataset that displays the distribution
        of numerical data and highlights important summary statistics.
        It provides a visual summary of the minimum, first quartile, median,
        third quartile, and maximum of a dataset, along with any potential outliers.

Box plots are used to understand the spread and skewness of the data, identify potential outliers,
and compare multiple datasets.
They are particularly useful when comparing the distribution of different groups or variables.
The key elements of a box plot are as follows:

The box represents the interquartile range (IQR),
which is the range between the first quartile (Q1) and the third quartile (Q3).
It encapsulates the middle 50% of the data.
The line within the box represents the median, which is the midpoint of the dataset when arranged in ascending order.
The whiskers extend from the box to the minimum and maximum values, excluding any outliers.
The length of the whiskers can vary depending on the data
and the definition used.
Outliers, if present, are plotted as individual points outside the whiskers.
To plot a box plot using the given data, you can use various software libraries or programming languages.
an example using Python and the Matplotlib library:


import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()

In this example, we create two sets of random data, box1 and box2, 
each containing 200 values generated from normal distributions
with different means and standard deviations. We then create a list data containing these two datasets.

Using plt.boxplot(data), we plot the box plot for the given data. 
The plt.xticks() function sets the labels for the x-axis to
'Box 1' and 'Box 2'. Finally, we set the y-axis label and the title of the plot using plt.ylabel()
and plt.title(), respectively. The plot is displayed using plt.show(). 
```
